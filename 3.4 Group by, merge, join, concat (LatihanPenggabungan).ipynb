```python
import pandas as pd
data = {'Perusahaan':['Babastudio','Babastudio','Techfor','Techfor'],
       'Nama':['Ahmad','Riyadli','Fahmi','techfor'],
       'Penjualan':[123,456,789,200]}
df = pd.DataFrame(data)
df
```

```python
#membuat group untuk kolom perusahaan
df.groupby('Perusahaan')
```

```python
#menampilkan rata-rata penjualan dari masing-masing perusahaan
df.groupby('Perusahaan').mean()
```

```python
#menampilkan data terkecil di masing-masing group
df.groupby('Perusahaan').min()
```

```python
#menampilkan data terbesar di masing-nasing perusahaan
df.groupby('Perusahaan').max()
```

```python
#menampilkan jumlah record disetiap perusahaan
df.groupby('Perusahaan').count()
```

```python
#menampilkan data-data group
df.groupby('Perusahaan').describe()
```

```python
#menampilkan data - data secara horizontal
df.groupby('Perusahaan').describe().transpose()
```

```python
df.groupby('Perusahaan').describe().transpose()['Babastudio']
```

```python
#menggabungkan 2 frame menjadi 1
left = pd.DataFrame({'Key1': ['K0','K0','K1','K2'],
                     'Key2': ['K0','K1','K0','K1'],
                        'A': ['A0','A1','A2','A3'],
                        'B': ['B0','B1','B2','B3']})

right = pd.DataFrame({'Key1': ['K0','K1','K1','K2'],
                      'Key2': ['K0','K0','K0','K0'],
                         'C': ['C0','C1','C2','C3'],
                         'D': ['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python

pd.merge(left, right, on=['Key1', 'Key2'])
```

```python
pd.merge(left, right, how='outer', on=['Key1','Key2'])
```

```python
pd.merge(left, right, how='right', on=['Key1', 'Key2'])
```

```python
pd.merge(left, right, how='left', on=['Key1','Key2'])
```

```python
#join untuk menggabungkan 2 dataframe menjadi 1
left = pd.DataFrame({'A': ['A0','A1','A2'],
                     'B': ['B0','B1','B2']},
                      index=['K0','K1','K2'])
right = pd.DataFrame({'C': ['C0','C2','C3'],
                      'D': ['D0','D2','D3']},
                      index=['K0','K2','K3'])
```

```python
left.join(right, how='inner')
```

```python
left.join(right, how='outer')
```

```python
left.join(right, how='left')
```

```python
left.join(right, how='right')
```
